# Predictive Analysing - Predict Life expectancy with Beer Wine Spirit Serving

```python
# code for buiding the model for predicting lidfe expectancy
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train = data_life_expect_at_birth.drop(['Life_Expectancy','GhoDisplay','GhoCode', 'RegionDisplay', 'YearCode','RegionDisplay', 'Income_Group', 'CountryDisplay','SexDisplay' ,'total_litres_of_pure_alcohol'], axis=1)
test = data_life_expect_at_birth['Life_Expectancy']
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=2)
regr = LinearRegression()
regr.fit(X_train, y_train)
pred = regr.predict(X_test)
print('Accuracy of the Model : ',regr.score(X_test, y_test))
print()
print(pred)
```
